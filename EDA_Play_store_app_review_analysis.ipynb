#### 3.1.2 Play store dataset - Handling NaN values

````python
# Handling the NaN values for the last two columns of Play store data

''' 
In 'Android Ver' column there are only 3 rows which contain NaN values and since NaN values in this column cannot be replaced by any particular value,
3 NaN value rows can be dropped.

 '''

ps_df = ps_df[ps_df['Android Ver'].notnull()]

# In 'Current Ver' column there are 8 NaN value rows and there is no particular value with which we can replace it, these rows can be dropped.

ps_df = ps_df[ps_df['Current Ver'].notnull()]
````

````python
# Finding corresponding value in 'Price' column where 'Type' column has single null value

ps_df[ps_df['Type'].isnull()]['Price'].values[0]
````

````python
# Since 'Price' column has value '0'(zero), NaN value in 'Type' column can be replaced by an equivalent term 'Free'

ps_df['Type'].fillna('Free', inplace = True)
````

````python
''' 
There are 1470 NaN values in the Rating column, thus it is not possible to remove these rows without losing a significant amount of data
that could affect the analysis's ultimate quality. 

'''

# The sum (mean or median) of the remaining values in the Rating column can be used in this situation to replace the NaN values.

mean_rating = round(ps_df[ps_df['Rating'].notnull()]['Rating'].mean(), 1)

median_rating = round(ps_df[ps_df['Rating'].notnull()]['Rating'].median(), 1)

print(f"The mean rating is {mean_rating} and median rating is {median_rating}")
````

````python
# Using a boxplot to identify outliers and a distplot to visualise the rating distribution

fig, ax = plt.subplots(2,1, figsize=(17,9))
sns.distplot(ps_df['Rating'],color='green',ax=ax[0])
sns.boxplot(x='Rating',data=ps_df, ax=ax[1])
plt.show()
````

````python
''' FINDINGS
(Without include the NaN values) The average ratings come to a mean of 4.2 and median value of 4.3.

Based on this, we may infer that 50% of the apps have an average rating above 4.3 while the other 50% have a rating below 4.3.

The ratings are left biassed, as seen in the distplot representations.

We are aware that if a variable is skewed, the values at the extreme ends of the distribution will affect the mean.
As a result, the median provides a more accurate depiction of the vast majority of values in the variable.

'''

# Therefore, we will use its median to replace the NaN values in the Rating column.

ps_df['Rating'].fillna(median_rating, inplace = True)
````

#### 3.1.4 Play store dataset - Changing the datatype of values in required columns

````python
# Changing the datatype of the 'Last Updated' column from string to datetime

ps_df['Last Updated'] = pd.to_datetime(ps_df['Last Updated'])
````

````python
# 'Price' column values are having '$' character at the begining

ps_df['Price'].value_counts()
````

````python
# Changing the datatype of the 'Price' column from string to float using lambda function 

# Using if else condition '$' character is removed through slicing

ps_df['Price'] = ps_df['Price'].apply(lambda x : float(x[1:]) if '$' in x else float(x))
````

````python
# 'Installs' column has '+' & ',' characters in its values

ps_df['Installs'].value_counts()
````

````python
# Changing the string datatype of the entries in the "Installs" column to an integer datatype

# Defining a function which removes '+' & ',' characters from string

def remove_plus_comma(install_string):
  '''
  This function removes '+' character through slicing and ',' character using replace() method and returns integer at the end
  '''
  if '+' and ',' in install_string:
    install_integer = int(install_string[:-1].replace(',',''))
    return install_integer
  elif '+' in install_string:
    install_integer_2 = int(install_string[:-1])
    return install_integer_2
  else:
    return int(install_string)

# Applying 'remove_plus_comma' function to the 'Installs' column

ps_df['Installs'] = ps_df['Installs'].apply(remove_plus_comma)
````

````python
# '+' & ',' characters are removed from column values

ps_df['Installs'].value_counts()
````

````python
# 'Size' column has values ending with 'M' & 'k' and also has string 'Varies with device'

ps_df['Size'].value_counts()
````

````python
ps_df[ps_df['Size'] == 'Varies with device']['Size']
````

````python
# Changing the string datatype of the values in the "Size" column to the same size measurement unit (MB)

# defining a function to convert all of the KB items to MB, then changing their datatype to float.

def from_kb_to_mb(size_string):
  '''
  The function returns the result as a float datatype after converting all of the valid KB elements to MB.
  Since 1MB = 1024KB, we can convert KB to MB by multiplying all of the KB numbers by 1024.
  '''
  if 'M' in size_string:
      return float(size_string[:-1])
  elif 'k' in size_string:
      return round(float(size_string[:-1])/1024, 4)
      
  # To return string as it is 'Varies with device' else part is added
  else:
      return size_string

# Applying 'from_kb_to_mb' function to 'Size' column values

ps_df['Size'] = ps_df['Size'].apply(from_kb_to_mb)
````

````python
# 'Size' column values are in mb size of measure in float datatype

ps_df['Size'].value_counts()
````

````python
''' 
Since the majority of entries in this column have the value "Varies with device," substituting it with any other value representing central tendency 
(such as mean or median) could provide inaccurate visualisations and results. 

'''

# So replacing 'Varies with device' string with float np.NaN

ps_df['Size'] = ps_df['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
ps_df['Size'] = ps_df['Size'].apply(lambda x: float(x))
````

#### Chart - 5 - New rating classification of the apps based on 'Rating' column.

````python
# Chart - 5 visualization code

# Since most of apps rating are above 4, considering median value rating for better analysis

median_rating = ps_df['Rating'].median()

# Defining a function group_app_rating to group the ratings

def group_app_rating(rating):
  ''''
  This function help to categories the rating from 1 to 5
  as Top_rated,Above_average,Average & below Average
  '''
  if rating >= median_rating:
    return 'Top rated'
  elif rating > 3 and rating < median_rating:
    return 'Above Average'
  elif rating > 2 and rating < 3:
    return 'Average'
  else:
    return 'Below Average'

# Applying group_app_rating function to 'Rating' column and creating new column 'Classify_rating'

ps_df['Classify_rating'] = ps_df['Rating'].apply(lambda x: group_app_rating(x))
````

#### Chart - 13 - Characteristic distribution of apps based on its size to find average number of installs and reviews of apps.

````python
# Chart - 13 visualization code

# Finding max, min, mean, and median in the Size column excluding the 'Varies with device' values.

max_size = ps_df[ps_df['Size'] != 'Varies with device']['Size'].max()

min_size = ps_df[ps_df['Size'] != 'Varies with device']['Size'].min()

mean_size = round(ps_df[ps_df['Size'] != 'Varies with device']['Size'].mean(),4)

median_size = ps_df[ps_df['Size'] != 'Varies with device']['Size'].median()

print(f"The calculated maximum, minimum, mean and median app size are {max_size}, {min_size}, {mean_size} and {median_size} MB respectively.")
````

````python
# Let us define a function to group the apps based on its size in MB

def size_apps(var):
  '''
  This function groups the size of an app 
  between ~0 to 100 MB into certain intervals.
  '''
  try:
    if var < 1:
      return 'Below 1'
    elif var >= 1 and var <10:
      return '1-10'
    elif var >= 10 and var <20:
      return '10-20'
    elif var >= 20 and var <30:
      return '20-30'
    elif var >= 30 and var <40:
      return '30-40'
    elif var >= 40 and var <50:
      return '40-50'
    elif var >= 50 and var <60:
      return '50-60'
    elif var >= 60 and var <70:
      return '60-70'
    elif var >= 70 and var <80:
      return '70-80'
    elif var >= 80 and var <90:
      return '80-90'
    else:
      return '90 and above'
  except:
    return var

# Lets apply the size_group function on the Size column and store the results in a new column named App Size group.

ps_df['App_size_group'] = ps_df['Size'].apply(lambda x : size_apps(x))
ps_df.head()
````
